# 02 Verify and compare


This notebook verifies a run and computes a diff between two manifests.

```
import json
from pathlib import Path
from omphalos.core.verify import verify_run_dir

run_dir = Path('artifacts/runs/demo')
report = verify_run_dir(run_dir)
print(report['ok'], report['files_checked'])
```

```
a = json.loads((Path('artifacts/runs/demo')/'manifest.json').read_text())
b = json.loads((Path('artifacts/runs/demo_old')/'manifest.json').read_text())

sa = {x['path']: x['sha256'] for x in a['artifacts']}
sb = {x['path']: x['sha256'] for x in b['artifacts']}
changed = sorted([p for p in set(sa)|set(sb) if sa.get(p)!=sb.get(p)])
print('changed', len(changed))
print(changed[:20])
```
