The command prints the run directory path. Copy it here (or re-run the cell).

```python
import glob, os
runs = sorted(glob.glob('artifacts/runs/run-*'))
run_dir = runs[-1] if runs else None
run_dir
```

```python
!python -m omphalos verify --run-dir {run_dir}
```

### Inspect the warehouse

The reference pipeline writes a SQLite file under `warehouse/warehouse.sqlite`.

```python
import sqlite3, pandas as pd
conn = sqlite3.connect(os.path.join(run_dir,'warehouse','warehouse.sqlite'))

pd.read_sql_query('select * from entity_scores order by chokepoint_score desc limit 10', conn)
```

### Inspect exports

Exports are under `exports/` and are intended for readers.

```python
import os
for root,_,files in os.walk(os.path.join(run_dir,'exports')):
    for f in sorted(files)[:10]:
        print(os.path.relpath(os.path.join(root,f), run_dir))
```
