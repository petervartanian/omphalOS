# 00 Run walkthrough


This notebook executes a reference run and inspects the resulting run directory.

```
import json, os
from pathlib import Path

root = Path('artifacts')
print('artifacts exists', root.exists())
```

```
from omphalos.core.runner import run_pipeline

run_id = 'demo_notebook'
run_dir = run_pipeline(run_id=run_id)
print(run_dir)
```

```
manifest = json.loads((Path(run_dir)/'manifest.json').read_text())
print(manifest['run_id'], len(manifest['artifacts']))
```

Inspect the top declared artifacts:

```
for a in manifest['artifacts'][:20]:
    print(a['path'], a['size'])
```
