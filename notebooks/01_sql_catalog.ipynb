# 01 SQL catalog usage


This notebook demonstrates running a small catalog manifest against the run warehouse.

```
import sqlite3, yaml
from pathlib import Path

run_dir = Path('artifacts/runs/demo')
warehouse = run_dir/'warehouse/warehouse.sqlite'
print('warehouse', warehouse.exists())
```

```
manifest = yaml.safe_load((Path('sql/manifests/briefing.yaml')).read_text())
print(manifest['name'], len(manifest['queries']))
```

```
conn = sqlite3.connect(str(warehouse))
for q in manifest['queries']:
    sql = Path(q['path']).read_text()
    cur = conn.execute(sql, {'limit': 20})
    rows = cur.fetchmany(5)
    print(q['path'], rows[:1])
conn.close()
```
